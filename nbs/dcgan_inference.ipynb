```python
import torch
import matplotlib.pyplot as plt
import numpy as np
from dcgan import GeneratorNew, DiscriminatorNew
from inference import make_grid
from utils import show_anim, show_batch
```

```python
gen_path = "models/dcgan/out/generator_size_64_fs_128_2.pth"
dis_path = "models/dcgan/out/discriminator_size_64_fs_64_2.pth"
FIXED_NOISE = torch.randn(64, 100, device="cuda")
```

```python

generator = GeneratorNew().to("cuda")
generator.load_state_dict(torch.load(gen_path))
generator.eval()

discriminator = DiscriminatorNew().to("cuda")
discriminator.load_state_dict(torch.load(dis_path))
discriminator.eval();

```

```python
with torch.no_grad():
    out = generator(FIXED_NOISE)
    outD = discriminator(out)

#out2 = get_output(generator, FIXED_NOISE)
```
